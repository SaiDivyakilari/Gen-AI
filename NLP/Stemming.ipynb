```python
words = [
    "running", "runner", "runs",
    "easily", "fairly", "happily",
    "studies", "studying", "studied",
    "flies", "flying", "flight",
    "argues", "argued", "arguing",
    "connection", "connections", "connected",
    "organization", "organized", "organizing",
    "management", "managed", "managing",
    "playing", "played", "player",
    "fishing", "fished", "fisherman",
    "caresses", "caressing", "caressed",
    "meeting", "meetings", "meetinghouse",
    "believable", "believability", "believe"
]
```

```python
from nltk.stem import PorterStemmer

stemming = PorterStemmer()

for word in words:
    print(f"{word} ---> {stemming.stem(word)}")

```

The meaning of the word is changing


```python
from nltk.stem import RegexpStemmer

reg_stem = RegexpStemmer('ing$|s$|able$|es$',min=4)

reg_stem.stem('eating')
```

```python

reg_stem = RegexpStemmer('^ing|^s|^able|^es',min=4)

reg_stem.stem('ingeating')
```

```python

reg_stem = RegexpStemmer('ing|s|able|es',min=4)

reg_stem.stem('eatingesing')
```

```python
from nltk.stem import SnowballStemmer

s_stem = SnowballStemmer('english')
for word in words:
    print(f"{word} --> {s_stem.stem(word)}")
```

Better than PorterStemmer
