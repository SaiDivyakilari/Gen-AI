```python
from nltk.tokenize import sent_tokenize , word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer, SnowballStemmer
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
```

```python
corpus = """
Natural Language Processing (NLP) is a fascinating field of Artificial Intelligence that focuses on the interaction between computers and humans through language.
It involves teaching machines how to understand, interpret, and generate human language in a valuable way.

Many NLP applications, such as sentiment analysis, machine translation, and speech recognition, rely on text preprocessing steps like tokenization, stemming, and lemmatization.
For example, a chatbot uses NLP techniques to understand what a user says and respond appropriately.

Researchers and engineers constantly improve NLP algorithms to make them more efficient and accurate. 
Recent advancements in deep learning and transformers have greatly improved the performance of language models.
"""
```

```python
stemmer = PorterStemmer()

sentences = sent_tokenize(corpus)

for i in range(len(sentences)):
    words = word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in stopwords.words('english')]
    sentences[i] = " ".join(words)
sentences
```

```python
stemmer = SnowballStemmer('english')

sentences = sent_tokenize(corpus)

for i in range(len(sentences)):
    words = word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in stopwords.words('english')]
    sentences[i] = " ".join(words)
sentences
```

```python
lemmatizer = WordNetLemmatizer()

sentences = sent_tokenize(corpus)

for i in range(len(sentences)):
    words = word_tokenize(sentences[i])
    words = [lemmatizer.lemmatize(word.lower()) for word in words if word not in stopwords.words('english')]
    sentences[i] = " ".join(words)
sentences
```
