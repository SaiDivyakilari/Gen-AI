```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
vocab_size = 10000
```

```python
one_hot_repr = [one_hot(words,vocab_size) for words in sentences]
one_hot_repr
```

```python
max_len = 0
for sent in sentences:
    word_count = len(sent.split())
    max_len = max(max_len,word_count)
max_len
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
```

```python
sent_len = 10
```

```python
embed_docs = pad_sequences(one_hot_repr,sent_len,padding='pre')
embed_docs
```

```python
from tensorflow.keras.models import Sequential

dimensions = 10     # feature representation
model = Sequential()
model.add(Embedding(vocab_size,dimensions,input_length = sent_len))
model.compile(optimizer='adam',loss='mse')
model.summary()
```

```python
model.predict(embed_docs)
```
