```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')


```

```python
def text_preprocessing(sentences):
    lemmatizer = WordNetLemmatizer()
    corpus = []
    for sent in sentences:
        cleaned = re.sub('[^a-zA-Z]',' ',sent)
        cleaned = cleaned.lower()
        words = cleaned.split()
        words = [lemmatizer.lemmatize(word) for word in words if word not in stopwords.words('english')]
        corpus.append(' '.join(words))
    return corpus


```

```python
X_train_prep = text_preprocessing(X_train['text'])
X_test_prep  = text_preprocessing(X_test['text'])
```

```python
from tensorflow.keras.preprocessing.text import one_hot,Tokenizer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
from tensorflow.keras.utils import pad_sequences
```

```python
#tokenizing the sentences
vocab_size = 10000

tokenizer = Tokenizer(num_words=vocab_size, oov_token='<OOV>')
tokenizer.fit_on_texts(X_train_prep)
```

```python
max_len = max(len(sent.split()) for sent in X_train_prep)
print(max_len)
```

```python

train_sequences = tokenizer.texts_to_sequences(X_train_prep)
test_sequences  = tokenizer.texts_to_sequences(X_test_prep)

X_train_pad = pad_sequences(train_sequences, max_len, padding='pre')
X_test_pad  = pad_sequences(test_sequences, max_len, padding='pre')
```

```python
dimensions = 10
model = Sequential()
model.add(Embedding(vocab_size,dimensions,input_length = max_len))
model.add(LSTM(64))
model.add(Dense(1,'sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
```

```python
model.fit(
    X_train_pad,
    y_train,
    validation_data=(X_test_pad, y_test),
    epochs=20,
    batch_size=4,
    callbacks=[early_stopping]
)
```
