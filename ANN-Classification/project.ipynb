```python
import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,StandardScaler,OneHotEncoder
```

```python
data = pd.read_csv("C:\\Users\\kilar\\OneDrive\\Desktop\\Gen AI\\ANN-Classification\\Churn_Modelling.csv")
data.head()
```

```python
#dropping unecessary values
data = data.drop(['RowNumber','CustomerId','Surname'],axis = 1)
data.head()
```

```python
# Label encoding
label_encoding = LabelEncoder()
data['Gender'] = label_encoding.fit_transform(data['Gender'])
data.head()
```

```python
# Encoding multi class variables

one_hot_encoding = OneHotEncoder()
encoded_df = one_hot_encoding.fit_transform(data[['Geography']])
encoded_df = encoded_df.toarray()
```

```python
features = one_hot_encoding.get_feature_names_out(['Geography'])
features
```

```python
geo_df = pd.DataFrame(encoded_df,columns=features)
geo_df
```

```python
data = data.drop(['Geography'],axis = 1)
data = pd.concat([data,geo_df],axis = 1)
data
```

```python
# tarin test split
X = data.drop(['Exited'],axis = 1)
y = data['Exited']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=23)
```

```python
#Scaling features

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

```

```python
# saving all the loaders as pickle files

import pickle
with open('label_encoding.pickle','wb') as file:
    pickle.dump(label_encoding,file)

with open('one_hot_encoding.pickle','wb') as file:
    pickle.dump(one_hot_encoding,file)

with open('Standard_Scaler.pickle','wb') as file:
    pickle.dump(sc,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
```

```python
model = Sequential()
model.add(Dense(64,activation = 'relu', input_shape = (X_train.shape[1],)))
model.add(Dense(32,activation = 'relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
from tensorflow.keras.optimizers import Adam
opt = Adam(learning_rate = 0.01)
model.compile(optimizer=opt,loss='binary_crossentropy',metrics = ['accuracy'])
```

```python
import datetime
log_dir = "logs/fit" + datetime.datetime.now().strftime("%Y%M%D-%H%M%S")
tensorflow_callback = TensorBoard(log_dir=log_dir)
```

```python
early_stopping = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=100,callbacks=[early_stopping,tensorflow_callback])
```

```python
model.save('model.h5')
```
